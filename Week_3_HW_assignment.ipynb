```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print(plt.style.available)
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use('ggplot')
```

```python
#label x as the range of values
x = np.linspace(-2, 4, 100)

#labely as the quadratic equation
y = x**2-2*x+1

#graph the function from the range of values
plt.plot(x,y)
plt.title("Quadratic function")
plt.ylabel('f(x)')
plt.xlabel('range of values')
```

```python
#defining the x values for the first two graphs and the third one for log(x) for x>0

x1 = np.linspace(-2,2,100)
x2 = np.linspace(0.01, 3, 100)

#Assign variable names to each graph
f = x1**3
g = np.exp(x1)
h = np.log(x2)


#graphing each function with a label and a different color
plt.plot(x1, f, label = 'x^3 graph', color = 'blue')
plt.plot(x1, g, label = 'e^x graph', color = 'g')
plt.plot(x2 , h , label = 'log(x) graph', color = 'brown')

#labeling the graphs
plt.title('graph functions')
plt.xlabel('x')
plt.ylabel('function values')
plt.legend()
```

```python
#range of values
x1 = np.linspace(-2,2,100)

#alternative style being used.
#The new style expands the plot and its good for such wide ranging value functions
plt.style.use('ggplot')

#I chose these functions as I enjoy trigonometry problems like sin and tan
tan_value = np.tan(x1)
sin_x_value = np.sin(x1)/x1


#I have to make sure to handle what will happen when x = 0 as it will see it as 1
sin_x_value[np.isnan(sin_x_value)] = 1

#scale of the function
plt.figure(figsize = (10,6))

#plotting the graph with different colors and linestyles
plt.plot(x1, tan_value, label = 'tan(x) graph', color = 'blue', linestyle = '--')
plt.plot(x1, sin_x_value, label = 'sin(x)/x graph', color = '0.25', linestyle = '-.')

#labels of functions and the x and y limits
plt.xlim(-2,2)
plt.ylim(-2,2)
plt.xlabel('x values')
plt.ylabel('Function values')
plt.title('graphed functions: tan(x) and sin(x)/x')
plt.legend()
```
