```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")
```

```python
Data = { 'Offense': ['Aggravated Assault', 'Burglary', 'Larceny', 'Motor Vehicle Theft', 'Murder', 'Rape', 'Robbery'],
    'Offenses per 100,000 people (2023)': [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5],
    'Change in Offenses per 100,000 People (2022-2023)': [-3.3, -8.1, -4.9, 12.0, -12.0, -9.8, -0.8]}

# Create the bar chart

bright_colors = ['#39FF14', '#FF3D00', '#00BFFF', '#FF00FF', '#FFD700', '#FF4500', '#7FFF00']

plt.figure(figsize=(10, 6))  # Set the figure size
plt.bar(Data['Offense'], Data['Offenses per 100,000 people (2023)'], color=bright_colors)

# Add x and y labels and a title
plt.xlabel('Offenses Committed')
plt.ylabel('Offenses per 100,000 People')
plt.title('Crime Rates for 2023')

# Rotate x-axis labels for better readability
plt.xticks(rotation=55)

plt.grid(axis = 'y', linestyle = '-.', color = 'Black')

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
plt.figure(figsize = (10,5))

Dark_colors = ['#0A0A0A', '#555D50', '#36454F', '#4A4A4A', '#353839', '#353E43', '#023020']

plt.barh(Data['Offense'], Data['Change in Offenses per 100,000 People (2022-2023)'], color = Dark_colors)




plt.xlabel('Change in offenses per 100,000')
plt.ylabel('Offenses committed')
plt.title('Change in crime rates from 2022-2023')

plt.grid(axis = 'y', linestyle = '--', color = 'black')

plt.tight_layout()

```
