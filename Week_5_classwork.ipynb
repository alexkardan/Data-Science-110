```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
titanic_df=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/refs/heads/main/dataset/titanic.csv')
```

```python
plt.style.use('ggplot')
plt.hist(titanic_df['age'], bins = 15, edgecolor = 'white')
plt.grid(axis = 'x')
plt.xticks([0,20,40,60])
plt.yticks([0,30,60,90,120])
plt.show()
```

```python

plt.figure(figsize = (10,6))
plt.hist(titanic_df['age'], bins = 15, color = 'skyblue', edgecolor = 'white')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age(years)')
plt.ylabel('count')
plt.grid(axis = 'x')
plt.xticks([0,20,40,60])
plt.tick_params(axis = 'x', colors = 'gray') # Set x-axis tick colors to gray
plt.show()
```

```python
ages_male = titanic_df[titanic_df['sex'] == 'male']['age']

ages_female = titanic_df[titanic_df['sex'] == 'female']['age']


plt.figure(figsize=(10,6))
plt.hist([ages_male,ages_female])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist([ages_male,ages_female], bins=15, label = ['Male', 'female'])

plt.title('Age Distribution by Sex on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist([ages_male,ages_female],stacked=True, edgecolor = 'white', bins=15, label = ['Male', 'female'])

plt.title('Age Distribution by Sex on the Titanic')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend()


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.hist(ages_male, bins = 20, label = 'Male', alpha = 0.5, color = 'blue', orientation = 'vertical', weights = [-1] *len(ages_male))

plt.hist(ages_female, bins = 20, label = 'Female', alpha = 0.5, color = 'red', orientation = 'vertical')

plt.xlabel('Number of Passengers')
plt.ylabel('Age')
plt.title('Age Pyramid of Titanic Passengers')


plt.yticks([-60,-40,-20,0,20,40,60], [60,40,20,0,20,40,60])

plt.legend(loc='upper left')

plt.show()
```

```python
plt.figure(figsize = (10,6))

plt.hist(ages_male, bins = 20, label = 'Male', alpha = 0.5, color = 'blue', orientation = 'horizontal', weights = [-1] *len(ages_male))

plt.hist(ages_female, bins = 20, label = 'Female', alpha = 0.5, color = 'red', orientation = 'horizontal')

plt.xlabel('Number of Passengers')
plt.ylabel('Age')
plt.title('Age Pyramid of Titanic Passengers')


plt.xticks([-80,-60,-40,-20,0,20,40,60,80], [80,60,40,20,0,20,40,60,80])

plt.legend(loc='upper left')

plt.show()
```

```python
plt.figure(figsize = (10,6))

sns.kdeplot(ages_male, fill = True, color = 'blue', label = 'Male')
sns.kdeplot(ages_female, fill = True, color = 'red', label = 'Female')
plt.legend()
plt.show()
```
