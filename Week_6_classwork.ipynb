```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from joypy import joyplot
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-22016/refs/heads/main/dataset/lincoln_temps.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.boxplot(x='month_long', y = 'mean_temp', data=df)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='month_long', y = 'mean_temp', data=df)
plt.xticks(rotation=45)
plt.show
```

```python
sns.boxplot(x='month_long', y = 'mean_temp', data=df, color = "skyblue")
plt.xticks(rotation=45)

plt.title('Mean Temperature by Month in Lincoln')
plt.xlabel('Month')
plt.ylabel('Mean Temperature')


plt.tight_layout
plt.show
```

```python
month_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
len(month_labels)
range(len(month_labels))
```

```python
sns.violinplot(x = 'month_long', y='mean_temp', data=df)
```

```python
sns.violinplot(x = 'month_long', y='mean_temp', data=df, inner = None)
```

```python
sns.violinplot(x='month_long', y = 'mean_temp', data=df, color = "skyblue")
plt.xticks(rotation=45)

plt.title('Mean Temperature by Month in Lincoln')
plt.xlabel('Month')
plt.ylabel('Mean Temperature')


plt.tight_layout
plt.show
```

```python
sns.stripplot(x='month_long', y='mean_temp', data=df)
plt.show()
```

```python
sns.stripplot(x='month_long', y = 'mean_temp', data=df, color = "skyblue",jitter=True)
plt.xticks(rotation=45)

plt.title('Mean Temperature by Month in Lincoln')
plt.xlabel('Month')
plt.ylabel('Mean Temperature')


plt.tight_layout
plt.show
```
