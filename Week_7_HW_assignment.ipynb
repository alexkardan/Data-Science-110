```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

iris.head()
```

```python
pair_plot = sns.pairplot(iris, hue = "species", markers = ["o", "s", "d" ])

pair_plot.fig.suptitle("Overall Species sepal length", y = 1.01)

plt.show()
```

The pair plot offers a comparison of feature distributions and relationships for each iris species. Key insights include that setosa species tend to cluster distinctly from versicolor and virginica, particularly in sepal and petal widths, indicating it has unique feature values. Meanwhile, versicolor and virginica show some overlap, especially in petal length and width, though virginica tends to have slightly larger measurements overall. This separation and overlap suggest potential for classification, with setosa being easily distinguishable and virginica and versicolor more closely related in terms of feature measurements.

```python
plt.figure(figsize = (8,5))
sns.barplot(data=iris, x = 'species', y = "sepal_length", errorbar = None )
plt.title('Average Sepal Length for every Iris species')
plt.xlabel('species')
plt.ylabel('average sepal length in cm')
plt.show()
```

The bar plot displays the average sepal length for each iris species, showing distinct differences among them. Virginica has the longest average sepal length, followed by versicolor, while setosa has the shortest. This trend suggests that sepal length can help differentiate setosa from the other species and provides some distinction between versicolor and virginica, supporting sepal length as a potentially valuable feature for classification.

```python
sns.boxplot(data=iris, x= "species", y = "sepal_length")
plt.title('Distribution of Petal widths for every Iris species')
plt.xlabel('species')
plt.ylabel('average petal width in cm')
plt.show()
```

This box plot shows the distribution of sepal length (not petal width) for each iris species, illustrating variations in sepal lengths within and between species. Setosa has a smaller and more tightly clustered range of sepal lengths, while versicolor and virginica exhibit greater variability, with virginica having the largest overall spread and higher values. This distribution indicates that sepal length variability could be useful in distinguishing setosa from the other two species, although there is some overlap between versicolor and virginica.








```python
sns.barplot(data=iris.melt(id_vars = "species", value_vars= ["sepal_length", "petal_length"]), x = "species", y = "value", hue = "variable", errorbar = None)
plt.title("Average Sepal and petal lengths for each species")
plt.xlabel("Species")
plt.ylabel("Length in cm")
plt.legend(title = "Measurment Type", labels = ["Sepal_length", "Petal_length"])
plt.show()
```

This bar plot illustrates the average lengths of sepals and petals for each iris species, providing a side-by-side comparison. Virginica exhibits the longest averages for both sepal and petal lengths, followed by versicolor, while setosa has the shortest averages. The significant differences in petal length between setosa and the other two species highlight petal length as a strong distinguishing feature. This visualization suggests that while both measurements vary across species, petal length may be particularly useful for differentiating setosa from versicolor and virginic

```python
sns.violinplot(data=iris, x="species", y="sepal_length")
plt.title("Distribution of Sepal Lengths for Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")
plt.show()
```

The violin plot displays the distribution and density of sepal lengths for each iris species, showing both the range and concentration of values. Setosa has a narrow range with a high density around lower sepal length values, while versicolor and virginica have broader distributions, with virginica reaching higher lengths overall. This visualization emphasizes that setosa consistently has shorter sepal lengths compared to the other species, with less overlap between setosa and the other two, making it relatively easy to distinguish. However, the overlap between versicolor and virginica suggests some similarity in sepal length, though virginica generally trends larger.
