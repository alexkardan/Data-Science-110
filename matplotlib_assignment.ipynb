```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
print(plt.style.available)
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use('ggplot')
```

```python
#label x as the range of values
x = np.linspace(-2, 4, 100)

#labely as the quadratic equation
y = x**2-2*x+1

#graph the function from the range of values
plt.plot(x,y)
plt.title("Quadratic function")
plt.ylabel('f(x)')
plt.xlabel('range of values')
```

```python
#defining the x values for the first two graphs and the third one for log(x) for x>0

x1 = np.linspace(-2,2,100)
x2 = np.linspace(0.01, 3, 100)

#Assign variable names to each graph
f = x1**3
g = np.exp(x1)
h = np.log(x2)


#graphing each function with a label and a different color
plt.plot(x1, f, label = 'x^3 graph', color = 'blue')
plt.plot(x1, g, label = 'e^x graph', color = 'g')
plt.plot(x2 , h , label = 'log(x) graph', color = 'brown')

#labeling the graphs
plt.title('graph functions')
plt.xlabel('x')
plt.ylabel('function values')
plt.legend()
```

```python
#range of values
x = np.linspace(-1,1,100)
#assighn each equation to a variable
F = x**2
G = x**3
H = x**4
I = x**5

#graph with 4 different specification methods and lebeled
plt.plot(x,F, label = 'x^2 graph', color = 'blue', linestyle = '-')
plt.plot(x,G,label = 'x^3 graph', color = 'g', linestyle = '--')
plt.plot(x,H, label = 'x^4 graph', color = '.35', linestyle = '-.')
plt.plot(x,I, label = 'x^5 graph', color = '#800080', linestyle = ':')

#labeling the graph
plt.title('Plot of polynomial functions')
plt.xlabel('x')
plt.ylabel('Function values')
plt.legend()
```
